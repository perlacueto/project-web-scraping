```python
import requests
from bs4 import BeautifulSoup
import time 
import re
import pandas as pd
```

```python
URL = 'https://olympics.com/tokyo-2020/olympic-games/en/results/all-sports/medal-standings.htm'
response = requests.get(URL)
content = response.content
soup = BeautifulSoup(content, 'html')

def scrap_pag_principal(soup):
    content_clean=soup.select('tr td.text-center,tr a.country')
    content_clean = content_clean[0:651] # eliminar datos de clasificacion por sexo
    content_clean = [content_clean[i].text for i in range(len(content_clean)) ]
    return content_clean

#myspider = Spider(URL, 1, content_parser=quotes_parser)

content_clean = scrap_pag_principal(soup)
```

```python
content_clean
```

```python
#Limpieza de datoss
content_clean = [re.sub(r'^ \r', '',dato) for dato in content_clean]
content_clean = [re.sub(r'^\r', '',dato) for dato in content_clean]
#content_clean = [re.sub(r' ', '_', content_clean[i]) for i in range(1,len(content_clean),7)]
#content_clean = [re.sub(r' ', '',dato) for dato in content_clean]
content_clean = [dato.strip() for dato in content_clean]
content_clean
```

```python
#Crea DataFrame
Posicion = [ int(content_clean[i]) for i in range(0,len(content_clean),7)]
Pais = [ content_clean[i] for i in range(1,len(content_clean),7)]
Oro = [ int(content_clean[i]) for i in range(2,len(content_clean),7)]
Plata = [ int(content_clean[i]) for i in range(3,len(content_clean),7)]
Bronce= [ int(content_clean[i]) for i in range(4,len(content_clean),7)]
Total_Medallas = [ int(content_clean[i]) for i in range(5,len(content_clean),7)]
Posicion_Total_Medallas =  [ int(content_clean[i]) for i in range(6,len(content_clean),7)]

dict2 = {'Pais':Pais,'Posicion':Posicion,'Oro':Oro,'Plata':Plata,'Bronce':Bronce,
        'Total_Medallas':Total_Medallas,'Posicion_Total_Medallas':Posicion_Total_Medallas}
data = pd.DataFrame(dict2)
```

```python
links_clean = [re.sub(r'^../../..', 'https://olympics.com/tokyo-2020/olympic-games',dato) for dato in links]
```

```python
links_clean
```

```python
def scrap_pag_seconds(url, country):
    
    response = requests.get(links_clean[0])
    content = response.content
    soup = BeautifulSoup(content, 'html')
    content_clean=soup.select('td a, tr td[class="StyleCenter"]')
    content_clean = [content_clean[i].text for i in range(len(content_clean)) ]
    content_clean = [re.sub(r'^\r', '',dato) for dato in content_clean]

    medallas = soup.select('tr td.text-center')
    medallas = [re.findall(r'alt=..',str(dato)) for dato in medallas ]
    medallas = [re.findall(r'[0-9]?',str(dato)) for dato in medallas ]
    medallas = [int(i[7]) for i in medallas]
    posiciones = ['Oro', 'Plata','Bronce'] 
    medallas = [posiciones[i-1] for i in medallas]
    
    Nombre = [ content_clean[i] for i in range(0,len(content_clean),3)]
    Deporte = [ content_clean[i] for i in range(1,len(content_clean),3)]
    Prueba = [ content_clean[i] for i in range(2,len(content_clean),3)]
    Pais = [country]*len(medallas) #identificador de pais a cada ganador
    
    dict1 = {'Pais':Pais,'Nombre':Nombre, 'Deporte':Deporte, 'Prueba':Prueba, 'Medalla':medallas, }
    medallero = pd.DataFrame(dict1)
    return medallero
```

```python
medalleros = []
for i in range(0,5): #se puede hacer scrap para todos los links de todos los paises que ganaron medallas, 
                        #pero tarda un montón en cargar, como muestra tomé los primeros 5
                        # si se quiere scrapear todo usa variable nro_paises_con_medallas en vez de 5
    medalleros.append(scrap_pag_seconds(links_clean[i], data['Pais'][i]))# lista de medalleros, 
                                                                    #cada índice es el medallero de un pais
    time.sleep(2)
medalleros
```

```python
medallistas = pd.DataFrame()
for medallero in medalleros:
    medallistas = pd.concat([medallistas,medallero])
medallistas
```
